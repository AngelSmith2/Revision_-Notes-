# If Statements 

```python
if 10 > 5:
   print("10 greater than 5")

print("Program ended")
# You can use if statements to run code if a certain condition holds.
# Use with True statements (runs the codes that are true ) 
# condition = 10>5
```

```python
num = 12
if num > 5:
   print("Bigger than 5")
   if num <=47:
      print("Between 5 and 47")
# if statements can be nested, one inside the other.
# This means that the 'inner if statement' is the statement part of the 'outer if statement'
```

```python
x = 4
if x == 5:
   print("Yes")
else:
   print("No")
#An else statement follows an if statement, and contains code that is called when the if statement evaluates to False. 
```

```python
num = 7
if num == 5:
  print("Number is 5")
else: 
  if num == 11:
    print("Number is 11")
  else:
    if num == 7:
      print("Number is 7")
    else: 
        
      print("Number isn't 5, 11 or 7")
# You can chain if and else statements to determine which option in a series of possibilities is true.
```

```python
num = 7
if num == 5:
   print("Number is 5")
elif num == 11:
   print("Number is 11")
elif num == 7:
   print("Number is 7")
else:
   print("Number isn't 5, 11 or 7")
# elif (short for else if)
# shortcut to use when chaining if and else statements.
# A series of elif statements can have a final else block, which is called if none of the if or elif expressions is True.
```

# While Loops

```python
i = 1
while i<=5:
    print(i)
    i=i+1
print ("finished")
# While statements is similar to if statements
# While statements can be rum more than once
# The statement is repeatedly executed as long as the condition holds 
# Once it evaluates to False, the next section of code is executed. 
```

```python
while 1==1:
  print("In the loop") 
# Infinite loops never stop running if the condition remains True 
```

```python
i = 0
while 1==1:
  print(i)
  i = i + 1
  if i >= 5:
    print("Breaking")
    break
print("Finished")
# To end a while loop prematurely, the break statement can be used
# The break statement causes the loop to finish immediately
```

```python
# Bad coding 
i = 0
while True:
   i = i +1
   if i == 2:
      print("Skipping 2")
      continue
   if i == 5:
      print("Breaking")
      break
   print(i)
print("Finished")
# Another statement that can be used within loops is continue
# continue jumps back to the top of the loop
```

# List

```python
words = ["Hello", "world", "!"]
print(words[0])
print(words[1])
print(words[2])
# They are used to store an indexed list of items. 
# A list is created using square brackets with commas separating items.
# The certain item in the list can be accessed by using its index in square brackets
```

```python
empty_list = []
print(empty_list)
# An empty list is created with an empty pair of square brackets.
```

```python
number = 3
things = ["string", 0, [1, 2, number], 4.56]
print(things[1])
print(things[2])
print(things[2][2])
# A list will contain items of a single item type, but it is also possible to include several different types. 
# Lists can also be nested within other lists.
```

```python
str = "Hello world!"
print(str[6])
# strings, can be indexed like lists
# Indexing strings behaves as though you are indexing a list containing each character in the string. 
```

```python
nums = [1, 2, 3]
print(nums + [4, 5, 6])
print(nums * 3)
# Lists can be added and multiplied in the same way as strings.
```

### List Operations

```python
words = ["spam", "egg", "spam", "sausage"]
print("spam" in words)
print("egg" in words)
print("tomato" in words)
# To check if an item is in a list, the in operator can be used
# It returns True if the item occurs one or more times in the list, and False if it doesn't.
```

```python
nums = [10, 9, 8, 7, 6, 5]
nums[0] = nums[1] - 5
if 4 in nums:
  print(nums[3])
else:
  print(nums[4])
```

```python
nums = [1, 2, 3]
print(not 4 in nums)
print(4 not in nums)
print(not 3 in nums)
print(3 not in nums)
# To check if an item is not in a list, you can use the not operator
```

### List Function

```python
nums = [1, 2, 3]
nums.append(4)
# The dot before append is there because it is a method of the list class.
print(nums)
# Another way of altering lists is using the append method.
# This adds an item to the end of an existing list.
```

```python
nums = [1, 3, 5, 2, 4]
print(len(nums))
# To get the number of items in a list, you can use the len function.
# Unlike append, len is a normal function, rather than a method. 
# This means it is written before the list it is being called on, without a dot.
```

```python
words = ["Python", "fun"]
index = 1
words.insert(index, "is")
print(words)
# The insert method is similar to append, it allows you to insert a new item at any position in the list
```

```python
nums = [9, 8, 7, 6, 5]
nums.append(4)
nums.insert(2, 11)
print(len(nums))
```

```python
letters = ['p', 'q', 'r', 's', 'p', 'u']
print(letters.index('r'))
print(letters.index('p'))
print(letters.index('z'))
# The index method finds the first occurrence of a list item and returns its index
```

# Range

```python
numbers = list(range(10))
print(numbers)
# The range function creates a sequential list of numbers
```

```python
numbers = list(range(3, 8))
print(numbers)
print(range(20) == range(0, 20))
# If range is called with one argument, it produces an object with values from 0 to that argument. 
# If it is called with two arguments, it produces values from the first to the second.
```

```python
numbers = list(range(5, 20, 2))
print(numbers)
# range can have a third argument, which determines the interval of the sequence produced. 
# This third argument must be an integer.
```

# For loops

```python
words = ["hello", "world", "spam", "eggs"]
counter = 0
max_index = len(words) - 1

while counter <= max_index:
   word = words[counter]
   print(word + "!") # Adding the exclamation mark makes the code print the word + !
   counter = counter + 1 
# you need to perform code on each item in a list. This is called iteration.
# It can be accomplished with a while loop and a counter variable.
```

```python
words = ["hello", "world", "spam", "eggs"]
for word in words:
  print(word + "!")
# Iterating through a list using a while loop requires quite a lot of code
# The for loop as a shortcut that accomplishes the same thing.
```

```python
for i in range(5):
  print("hello!")
# The for loop is commonly used to repeat some code a certain number of times. 
# This is done by combining for loops with range objects.
```
